```python
# Load data
df = pd.read_csv('training_data.csv')

# Learn the structure (DAG)
DAG = bn.structure_learning.fit(df)

# Print adjacency matrix
print(DAG['adjmat'])

# Static matplotlib plot - Save as PNG
G_pyplot = bn.plot(DAG)
plt.savefig("bayesian_network_static.png", dpi=300)
plt.close()

# Interactive pyvis plot - Save as HTML
G_interactive = bn.plot(DAG, interactive=True)
G_interactive.save_graph("bayesian_network_interactive.html")

# Graphviz plot - Save as PDF
bn.structure_learning.plot(DAG, methodtype='graphviz', filename='bayesian_network_graphviz.pdf')

bn.plot_graphviz(DAG)
```
