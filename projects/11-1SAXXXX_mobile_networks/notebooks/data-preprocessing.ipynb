# Preprocessing #

First all rows are being removed that have unrealistic values for the target feature **NR SS-RSRP [dBm]**. This will unfortunately remove around 4000 entries in the dataframe and reduce the available data by roughly 30%.

```python
import pandas as pd
import matplotlib.pyplot as plt

with open('../data/measurement_data_export.csv', 'r') as file:
    df = pd.read_csv(file, sep=',', header=0)
    
df = df[df['NR SS-RSRP [dBm]'] < -10]

df.describe()
```

```python
df.to_csv('../data/cleaned_data_RSRP.csv', index=False)
```
